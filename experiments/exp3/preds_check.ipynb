```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('preds_2000_2002.csv')
```

```python
df = df.set_index(['level_0', 'lon', 'lat'])[['pred0','pred_low0', 'y_var0', 'y_var_low0']]
```

```python
da = df.to_xarray()
```

```python
da.isel(level_0=12).pred0.plot(x='lon',y='lat', figsize=(10,5), vmin=-4, vmax=4)
```

```python
da.isel(level_0=12).pred_low0.plot(x='lon',y='lat', figsize=(10,5), vmin=-4, vmax=4)
```
